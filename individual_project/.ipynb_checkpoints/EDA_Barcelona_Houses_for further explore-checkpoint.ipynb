def _transform_column_into_dummies(dataframe, name_of_column):
    dummies = pd.get_dummies(dataframe[name_of_column], prefix=f"{name_of_column} = ")
    dataframe = pd.concat([dataframe, dummies], axis=1)
    dataframe = dataframe.drop([name_of_column], axis=1)
    return dataframe

#### Now we would like to explore the investment value on Real Estate in Barcelona.

We have now the cleaned dataset of 'Average offer price in the neighborhoods (€:m2). Annual data 2013-2023.csv' and 'Barcelona_rent_price_2014-2022.csv'. Let's combine these 2 dataset from where we could calculate the return rate using the data of selling price and renting price per surface.

Firstly, we have to make some processing and combine the 2 datasets.

```python
data_price_m2_district
```

```python
data_district_price_per_m2 = data_price_m2_district.drop(columns=['District_num']) 
```

```python
# We want to know the average selling price per surface of different districts from 2013 to 2023.
district_price_per_m2 = data_district_price_per_m2.groupby(['District', 'Year'])['Price/m2'].mean().round(2).reset_index()
```

```python
district_price_per_m2.head(20)
```

```python
# As we want to see the price of each districts, we would drop the rows of data of Barcelona.
district_price_per_m2.drop(index=range(0,11), inplace=True)
```

```python
# As there is no renting data of the year of 2013 and 2023, we won't consider the data of these 2 years. So we have to filter the selling price data without data of 2013 and 2023.
mask = (district_price_per_m2['Year']!= '2013') & (district_price_per_m2['Year']!= '2023')
```

```python
# Now we have the df1 which is selling price per surface from 2014 to 2022.
df1 = district_price_per_m2[mask]
```

```python
df1 = df1.reset_index(drop=True)
```

```python
df1
```

```python
df1.dtypes
```

```python
df1['Year']= pd.to_numeric(df1['Year'])
```

```python
df1.dtypes
```

```python
# Let's look at the renting price per surface.
data_rent_district_per_m2
```

```python
# Now we have the df2 which is renting price per surface from 2014 to 2022.
df2 = data_rent_district_per_m2.copy()
```

```python
df2 = df2.rename(columns={'Price': 'Renting_price/m2'})
```

```python
df2
```

```python
df2 = df2.sort_values(by=['District', 'Year'], ascending=True)
```

```python
df2 = df2.reset_index(drop=True)
```

```python
df2
```

```python
df2.dtypes
```

Now we have to combine these 2 dataframes.
* When I firstly tried using pd.merge(), the combined dataframe contain a lot of NaN value. So I returned to check the following aspects:
1) if the data types are the same regarding the shared columns.
2) if there is whitespace or case incoincident problem
3) if values of 'District' column are not coincident

```python
df1['District'].unique()
```

```python
df2['District'].unique()
```

```python
# Rename district names using the dictionary
rename_dict = {'Gracia': 'Gràcia', 'Horta-Guinardo': 'Horta-Guinardó', 'Sant Marti': 'Sant Martí', 'Sants-Montjuic': 'Sants-Montjuïc', 'Sarria-Sant Gervasi': 'Sarrià-Sant Gervasi'}

df2['District'] = df2['District'].replace(rename_dict)
```

```python
# To eliminate whitespace
df1['District'] = df1['District'].str.strip()
df2['District'] = df2['District'].str.strip()
```

```python
# Ensure the 'District' columns have the same case
# df1['District'] = df1['District'].str.upper()
# df2['District'] = df2['District'].str.upper()
```

```python
df = pd.merge(df1, df2, on=['Year', 'District'], how='inner')
```

```python
df
```

```python
df['return_rate'] = df['Renting_price/m2'] / df['Price/m2']*100
```

```python
df
```

```python
plt.figure(figsize=(10,6))

sns.lineplot(
    data=df, 
    x="Year", 
    y="return_rate",
    hue='District'
)

plt.title('Return rate of properties in different districts in Barcelona 2014-2022')
```

From above line plot we can see that the general trend of investment return considering the renting income and selling price, fluctuate constantly. There are some peaks like in 2015 and 2020 respectively and also some sharp drops like 2018 and 2021. Think of the plots of selling prices trend and renting prices trend we've seen before, there are some points which draw the attentions:

1) The return rate of Nou Barris ranks at the top. The selling price of this district is at the bottom, so as the renting price. But as the And it was in dropping trend from 2017 to 2019, but it returned to go up since then and it was rarely affected by Covid Pandemia and the return keeps an comparatively stable status. It might because the renting price is increasing faster than the selling price and the district is preferred district in renting market. We could say that this is a district worth to invest.
2) The trend of Gracia, Sant Marti, Sants-Monjuic and Sant Andreu are continously increasing after Pandemia, considering the value of return rate, Gracia could be the most worthy district to invest, then Sant Marti, Sants-Monjuic and Sant Andreu.
3) Ciutat Vella faced a sharp drop of return rate, combining with the previous data, this could because the selling price creases a lot but renting keeps at the previous level.
