#### Now we would like to explore the investment value on Real Estate in Barcelona regarding renting.

We have now the cleaned datasets of 'Average offer price in the neighborhoods and districts 2013-2023' and 'Barcelona_rent_price_2014-2022'. Let's combine these 2 datasets from where we could calculate the return rate using the data of selling price and renting price per surface.

Now we have to start from processing the 2 datasets before combining.

```python
data_price_m2_district
```

```python
data_district_price_per_m2 = data_price_m2_district.drop(columns=['District_num']) 
```

```python
# We want to know the average selling price per surface of different districts from 2013 to 2023.
district_price_per_m2 = data_district_price_per_m2.groupby(['District', 'Year'])['Price/m2'].mean().round(2).reset_index()
```

```python
district_price_per_m2.head(20)
```

```python
# As we want to see the price of each districts, we would drop the rows of data of Barcelona.
district_price_per_m2.drop(index=range(0,11), inplace=True)
```

```python
# As there is no renting data of the year of 2013 and 2023, we won't consider the data of these 2 years. So we have to filter the selling price data without data of 2013 and 2023.
mask = (district_price_per_m2['Year']!= '2013') & (district_price_per_m2['Year']!= '2023')
```

```python
# Now we have the df1 which is selling price per surface from 2014 to 2022.
df1 = district_price_per_m2[mask]
```

```python
df1 = df1.reset_index(drop=True)
```

```python
df1
```

```python
df1.dtypes
```

```python
df1['Year']= pd.to_numeric(df1['Year'])
```

```python
df1.dtypes
```

```python
# Let's look at the renting price per surface.
data_rent_district_per_m2
```

```python
# Now we have the df2 which is renting price per surface from 2014 to 2022.
df2 = data_rent_district_per_m2.copy()
```

```python
df2 = df2.rename(columns={'Price': 'Renting_price/m2'})
```

```python
df2
```

```python
df2 = df2.sort_values(by=['District', 'Year'], ascending=True)
```

```python
df2 = df2.reset_index(drop=True)
```

```python
df2
```

```python
df2.dtypes
```

Now we have to combine these 2 dataframes.
* When I firstly tried using pd.merge(), the combined dataframe contain a lot of NaN value. So I returned to check the following aspects:
1) if the data types are the same regarding the shared columns.
2) if there is whitespace or case incoincident problem
3) if values of 'District' column are not coincident

```python
df1['District'].unique()
```

```python
df2['District'].unique()
```

```python
# Rename district names using the dictionary
rename_dict = {'Gracia': 'Gràcia', 'Horta-Guinardo': 'Horta-Guinardó', 'Sant Marti': 'Sant Martí', 'Sants-Montjuic': 'Sants-Montjuïc', 'Sarria-Sant Gervasi': 'Sarrià-Sant Gervasi'}

df2['District'] = df2['District'].replace(rename_dict)
```

```python
# To eliminate whitespace
df1['District'] = df1['District'].str.strip()
df2['District'] = df2['District'].str.strip()
```

```python
# Ensure the 'District' columns have the same case
# df1['District'] = df1['District'].str.upper()
# df2['District'] = df2['District'].str.upper()
```

```python
df = pd.merge(df1, df2, on=['Year', 'District'], how='inner')
```

```python
df
```

```python
# The surface of properties should be considered when we calculate the return rate of investment using the value per surface. 
# But as we didn't have that value in the dataset we are working with, and this value could be offset during the calculation process, we ignore this value this time.
df['Return_rate'] = df['Renting_price/m2'] / df['Price/m2']*100
```

```python
df
```

```python
plt.figure(figsize=(10,6))

sns.lineplot(
    data=df, 
    x="Year", 
    y="Return_rate",
    hue='District'
)

plt.title('Renting return rate of properties in different districts in Barcelona 2014-2022')
```

From above lineplot we can see that the general trend of investment return considering the renting income and selling price fluctuated constantly. There are some peaks like in 2015 and 2020 respectively and also some sharp drops like 2018 and 2021. Think of the plots of selling prices trend and renting prices trend we've seen before, there are some points where draw attentions:

1) The return rate of Nou Barris ranks at the top. Although the selling price of this district is at the bottom, so as the renting price, the return_rate is comparatively attractive. And the general trend of the return rate is increasing. Even the selling price of this district is low, but there is no extremely big gap of renting price comparing with other districts, that might be the reason that its return rate is the highest. So we could consider to invest in this area if we would like to buy properties for renting.
2) The renting return rate of the district Ciutat Vella rapidly increased since 2021. Combining what we've seen before, the selling price doesn't rank on the top and it increased smoothly, but the rentng price is sharply increased, which made the renting return rate increase a lot. Regarding this, we could consider this district as potential district to invest for renting.
3) Regarding the districts we generally considered as "traditional best districts" as Sarria-Sant Gervasi, Lets Corts and Eixample, actually their renting return rate is not very high. The reason is that the selling price of the properties is very high, but the renting price comparing with other districts didn't go obviously up. So if we consider investing and getting income through renting properties, these areas might not be the top options.

## Conclusion

If we would like to invest in properties in Barcelona and get profits through renting, and we pursue high investment return rate, the districts as Nou Barris, Horta-Guinardo might be good options. Except for that, Ciutat Vella might be the very potential district for investment for renting.
